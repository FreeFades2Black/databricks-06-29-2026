## Agent Components and Use Cases

### AI Agent Systems
**Non-agentic (static) workflow**
- Static prompt response or fixed pipelines of hardcoded prompt response systems
- Could also be an API call to another service or ML model.
- Deterministic actions

**Agentic (dynamic, iterative) workflow**
- Planning and execution by AI
- Tool calling by AI
- Non-deterministic actions
- Iterative workflows

![agentvsnon-agent.png](../Includes/images/agentvsnon-agent.png "agentvsnon-agent.png")

### Components of Agentic Systems
**LLM**
- A “brain”--LLM to control the core logic and sequencing of actions the agent executes. It plans the work of the interacting components.
- Other LLMs or AI models as needed for sub-tasks and actions.

**Tools**

External resources that the agent uses via tool use or tool calling, e.g., functions, APIs, classic ML models.

**Memory**
- Short-term session and current conversational state tracking to help with planning and execution of subsequent actions.
- Long-term episodic, semantic, and procedural memory for historical state, knowledge, preferences. 

### Common Patterns
####Pattern: Reasoning
**Chain-of-Thought**
A prompting technique that can be used to illicit reasoning from an instruct model.

**Reasoning model**
A model that has been trained to automatically determine when to think about a problem.

#### Pattern: ReAct (Reason + Act)
- Enables models to generate verbal reasoning traces and actions.
- Main states used in a ReAct agent are;
  - **Thought**: Reflect on the problem given and previous actions taken
  - **Act**: Choose the correct tool and input format to use.
  - **Observe**: Evaluate the result of the action and generate next thought.

#### Pattern: Plan-And-Solve
- Enables models to generate plans and then execute and observe.
- Main components are
  - **Planner**: Reflect on the task and generate a series of steps to complete the task
  - **Execute**: Process step(s) in the generated plan using tools and observations

### Use Cases
![use-cases.png](../Includes/images/use-cases.png "use-cases.png")

**Intelligent Document Processing**

With an agentic system, we can extract deep, contextual insights from massive volumes of documents at scale. Think of an agent that doesn't just 'read' a contract, but cross-references clauses against company policy and flags discrepancies automatically.

**Knowledge Base + Search**

AI agents transform search systems into dynamic retrieval systems. Instead of giving you a list of links, these agents search through your internal data, synthesize the information, and provide a direct, cited answer. It's the difference between finding a document and finding the actual answer hidden within it.

**Machine Learning + AI**

We use classic ML for what it does best—-like predictive analytics and structured data forecasting-—and then use Generative AI agents to interpret those results. They can explain the 'why' to a human user and suggest the next best action.
