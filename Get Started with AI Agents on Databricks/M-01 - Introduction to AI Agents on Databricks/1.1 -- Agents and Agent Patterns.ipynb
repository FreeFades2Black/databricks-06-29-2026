# Agents and Agent Patterns

## What is an Agent?
An AI agent is an intelligent application that uses an AI model and other tools to iteratively plan and execute sequences of actions to complete a complex task.
- Agents automate processes that previously only humans could do
- Agents do this by prompting an LLM to reason about what data or APIs or other Agents should be used to effectively respond to a user’s request
- Agent ROI is usually calculated by “# of human hours saved”

## Fitting AI Agents into Broader AI Systems
**AI systems** - encompass any software that utilizes machine learning models to function.

**Compound AI Systems** - interact with external components—-like databases, search engines, or other APIs—-to deliver a more reliable and factual output.

**Agentic AI** - the environment or the strategy that orchestrates agent workers. When we talk about Agentic AI, we are usually referring to Multi-Agent Systems. This is where multiple specialized agents collaborate, peer-review each other, and execute complex workflows without constant human hand-holding.

**AI Agents** - the individual 'workers.' They are defined by four specific capabilities: they have an AI model as a brain, planning capabilities to break down tasks, memory to track progress, and the ability to use tools (like running code or searching a file).

## Multi-Agent Systems
In a multi-agent system, a type of Agent, called a supervisor (or moderator/ or router) combines multiple Agents together.  

The supervisor:
- Uses an LLM to decide which agent should speak next
- Deterministically passes the conversation history to that agent (e.g., via code, no LLM regurgitating the tokens)
- Lets that agent respond
- Based on that response + conversation history, decides to either loop through the process again or return the output to the user

## Definitions
**Conversational Agents**: A conversational agent (chat agent, chatbot) is designed for iterative collaboration with humans and other agents through multi-turn conversations, where the conversation history is maintained somewhere. They don't just talk; they reason and act based on a continuous stream of context. They transform a simple text box into a functional interface for complex tasks.

**Tool**: A specific function that an AI model uses to perform a task. The details needed for the function (called parameters) are selected by the AI itself. A tool is designed to complete a specific task in a single step.

**Tool Calling Agent**: (aka, "Function Calling Agent") An Agent that has a set of Tools.  The Agent uses an LLM to:
- Reason about which Tool(s) to call
- Call those tools
- Reason about tool output and either call additional Tool(s) or respond to the user
