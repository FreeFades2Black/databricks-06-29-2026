# Demo - Querying the Agent Model
Now that we have deployed our model to Databricks Model Serving, we can query the model.
### Learning Objectives
_By the end of this demo, you will be able to:_ 
- View a deployed model serving endpoint
- Query a model serving endpoint using the UI
- Query a model serving endpoint using SQL

### Querying an Agent Deployed to Mosaic AI Model Serving Using the UI
The last cell in the `driver` notebook deployed our agent model to Databricks Model Serving. The deployment process takes about 10-12 minutes.

#### Instructions
1. Navigate to **Serving** on the left side menu and click on the endpoint called `agents_dbacademy-get_started_agents-my_first_agent`.
1. Click on **Use** at the top right. 
1. Begin querying! Here are some examples of queries you can use: 
    - _Get the order history of ronald54@example.net_
    - _Can you tell me the policy for exchanging items?_

> Please note: If you are using Databricks Free Edition, the endpoint may have scaled to zero. This means the query will take longer than normal.

### Querying an Agent Deployed to Mosaic AI Model Serving Using SQL
We can also query the model using SQL. We will use the built-in function, called [ai_query()](https://docs.databricks.com/aws/en/sql/language-manual/functions/ai_query) to query the model, passing the endpoint name as the first parameter.

```sql
SELECT ai_query(
  'agents_dbacademy-get_started_agents-my_first_agent',
  '{"input": [{"role": "user", "content": "Can you tell me the policy for exchanging items?"}]}'
) AS Output
```
